```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
%matplotlib inline
```

```python
data_dir = Path('/mnt/c/Users/Ronan Perry/Documents/JHU/jovo-lab/multiview/data')
embed_dir = data_dir / 'embedding'
```

```python
files = ['fmri_gcca-n=10_315-10_52.csv', 'sc_gcca-n=10_315-10_52.csv', 'goa_gcca-n=10_315-10_52.csv']
```

```python
Xs = [np.genfromtxt(embed_dir / file, delimiter=' ') for file in files]
```

```python
plt.style.use('seaborn')

y = list(range(10))

color_map = [sns.color_palette("Set2", 10)[int(i)] for i in y]

fig, axes = plt.subplots(1,3,figsize=(18,5), sharey=True)

for i, (X_latent, ax) in enumerate(zip(Xs, axes.flatten())):
    ax.scatter(x=X_latent[:,0], y=X_latent[:,1], alpha=1, label=y, color=color_map, s=25)
    ax.set_title(f'View {i+1}')
    
ax = fig.add_subplot(111, frameon=False)
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
ax.grid(False)
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
```
